```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from grafo import Grafo
```

```python
# Criação do grafo
grafo = Grafo()

# Leitura do arquivo de entrada
# Substitua 'exemplo.txt' pelo caminho correto do arquivo
grafo.ler_arquivo('exemplo.txt')
```

```python
# Estatísticas do grafo
estatisticas = grafo.estatisticas()

# Componentes conectados
componentes = grafo.componentes_conectados()
```

```python
# Exibição de estatísticas no console
def exibir_estatisticas(estatisticas):
    print("\nEstatísticas do Grafo:")
    for chave, valor in estatisticas.items():
        print(f"{chave}: {valor}")

# Chamando a função de exibição
exibir_estatisticas(estatisticas)
```

```python
# Exibição das estatísticas em formato tabular (usando pandas)
def tabela_estatisticas(estatisticas):
    df = pd.DataFrame(list(estatisticas.items()), columns=["Estatística", "Valor"])
    print("\nTabela de Estatísticas:")
    print(df)

# Chamando a função de tabela
tabela_estatisticas(estatisticas)
```

```python
# Exibição dos componentes conectados
def exibir_componentes(componentes):
    print("\nComponentes conectados:")
    for i, componente in enumerate(componentes, start=1):
        print(f"Componente {i}: {componente}")

# Chamando a função de componentes conectados
exibir_componentes(componentes)
```

```python
# Visualização do grafo usando NetworkX
def visualizar_grafo(grafo):
    G = nx.Graph()

    # Adiciona vértices e arestas ao grafo de NetworkX
    for v in grafo.vertices:
        G.add_node(v)
    for u, v, peso in grafo.arestas:
        G.add_edge(u, v, weight=peso)

    # Configuração do layout e visualização
    pos = nx.spring_layout(G)
    plt.figure(figsize=(10, 7))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=600, font_size=10)
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title("Visualização do Grafo")
    plt.show()

# Chamando a função para visualizar o grafo
visualizar_grafo(grafo)
```

```python
# Visualização de histogramas
def visualizar_histogramas(grafo):
    # Distribuição dos graus dos vértices
    graus = grafo.grau_vertices()
    plt.figure(figsize=(8, 5))
    plt.bar(graus.keys(), graus.values(), color='orange')
    plt.title("Distribuição dos Graus dos Vértices")
    plt.xlabel("Vértices")
    plt.ylabel("Grau")
    plt.show()

    # Intermediação (Betweenness Centrality)
    betweenness = grafo.calcular_intermediacao()
    plt.figure(figsize=(8, 5))
    plt.bar(betweenness.keys(), betweenness.values(), color='purple')
    plt.title("Intermediação (Betweenness Centrality)")
    plt.xlabel("Vértices")
    plt.ylabel("Intermediação")
    plt.show()

# Chamando a função para visualizar histogramas
visualizar_histogramas(grafo)
```
